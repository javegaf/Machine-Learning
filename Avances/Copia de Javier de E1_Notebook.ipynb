#### 🧭 **Definición de fases de la metodología Crisp-DM utilizada a lo largo del semestre**

- **Fase 1: Comprensión del negocio (Business Understanding)**

  En esta etapa, se busca comprender los objetivos y requisitos del negocio desde la perspectiva del problema que se está abordando. Se definen los objetivos del proyecto, se establece el marco de trabajo y se desarrolla un plan inicial para alcanzar esos objetivos.

- **Fase 2: Comprensión de los datos (Data Understanding)**

  Aquí, se recopila y se familiariza con los datos disponibles para el proyecto. Se realiza una exploración inicial de los datos para comprender su estructura, calidad, distribuciones y posibles relaciones entre las variables. Esto puede implicar la realización de análisis estadísticos descriptivos, visualización de datos y evaluación de la calidad de los datos.

- **Fase 3: Preparación de los datos (Data Preparation)**

  Durante esta etapa, se limpian y preparan los datos para el modelado. Esto puede incluir la selección de variables relevantes, la transformación de datos, la imputación de valores faltantes, la codificación de variables categóricas y la creación de nuevas características derivadas. El objetivo es garantizar que los datos estén en un formato adecuado y de alta calidad para el análisis posterior.

- **Fase 4: Modelamiento (Modeling)**

  En esta etapa, se seleccionan y se aplican técnicas de modelado para construir y evaluar modelos predictivos o descriptivos. Se pueden probar varias técnicas y algoritmos de modelado, y se ajustan los parámetros de los modelos para optimizar su rendimiento. Esta etapa también implica la evaluación y comparación de los modelos para seleccionar el más adecuado según los criterios definidos anteriormente.

- **Fase 5: Evaluación del modelo (Evaluation)**

  Aquí se evalúan los modelos desarrollados para determinar su idoneidad para abordar los objetivos del negocio. Se utilizan métricas de evaluación adecuadas para medir el rendimiento de los modelos, como precisión, recall, F1-score, entre otras. Además, se puede realizar una validación cruzada o dividir los datos en conjuntos de entrenamiento y prueba para evaluar el rendimiento del modelo en datos no vistos.

- **Fase 6: Despliegue (Deployment)**

  La etapa final implica implementar los resultados del proyecto en el entorno operativo del negocio. Esto puede implicar la integración de los modelos en sistemas existentes, la creación de informes o tableros interactivos, la automatización de procesos o cualquier otra acción necesaria para garantizar que los resultados se utilicen de manera efectiva para generar valor comercial.

**1. ¿Qué meses presentan mayor o menor facturación?**

**2. ¿Existe correlación entre la cantidad de transacciones y los pagos efectivos?**

**Objetivo:** Evaluar la relación entre las transacciones realizadas por los clientes y los pagos efectivos para ver si estos están directamente relacionados.

**Variables a analizar:**
- **Transacciones por mes:** Txs_T01, Txs_T02, ..., Txs_T12

- **Pagos por mes:** Pago_T01, Pago_T02, ..., Pago_T12

**Matriz de correlación**

Calcularemos la correlación entre las transacciones y los pagos para cada mes.

```python
# Definir columnas de transacciones y pagos reales
transacciones = ['Txs_T01', 'Txs_T02', 'Txs_T03', 'Txs_T04', 'Txs_T05', 'Txs_T06',
                 'Txs_T07', 'Txs_T08', 'Txs_T09', 'Txs_T10', 'Txs_T11', 'Txs_T12']

pagos_nacionales = ['PagoNac_T01', 'PagoNac_T02', 'PagoNac_T03', 'PagoNac_T04', 'PagoNac_T05', 'PagoNac_T06',
                    'PagoNac_T07', 'PagoNac_T08', 'PagoNac_T09', 'PagoNac_T10', 'PagoNac_T11', 'PagoNac_T12']

pagos_internacionales = ['PagoInt_T01', 'PagoInt_T02', 'PagoInt_T03', 'PagoInt_T04', 'PagoInt_T05', 'PagoInt_T06',
                         'PagoInt_T07', 'PagoInt_T08', 'PagoInt_T09', 'PagoInt_T10', 'PagoInt_T11', 'PagoInt_T12']

# Crear columnas de pagos totales por mes
for i in range(12):
    mes = f"T{str(i+1).zfill(2)}"
    data_frame[f'PagoTotal_{mes}'] = pd.to_numeric(data_frame[f'PagoNac_{mes}'], errors='coerce') + \
                                     pd.to_numeric(data_frame[f'PagoInt_{mes}'], errors='coerce')

# Convertimos las columnas de transacciones a numérico
data_frame[transacciones] = data_frame[transacciones].apply(pd.to_numeric, errors='coerce')

# Seleccionamos las nuevas columnas de pagos totales
pagos_totales = [f'PagoTotal_T{str(i+1).zfill(2)}' for i in range(12)]

# Calculamos la correlación entre transacciones y pagos
correlacion = data_frame[transacciones + pagos_totales].corr()

# Extraemos la submatriz de correlación entre transacciones y pagos
correlacion_tx_pago = correlacion.loc[transacciones, pagos_totales]

# Visualizamos la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlacion_tx_pago, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title('Correlación entre Transacciones y Pagos por Mes')
plt.xlabel('Pagos Totales')
plt.ylabel('Transacciones')
plt.show()
```

---

**Observaciones**

---

- La correlación más alta es de 0.29, lo que indica una relación ligeramente positiva, pero no fuerte.

- Las correlaciones están en torno a 0.26 - 0.29 en el centro del año, lo que sugiere que hay una cierta asociación en esos meses.

- No se observan correlaciones negativas fuertes, lo cual también es relevante.

**Conclusión**

Existe una correlación positiva débil entre la cantidad de transacciones y los pagos efectivos mensuales. Las correlaciones más altas (máximo de 0.29) se observan en los meses centrales del año, lo que sugiere una ligera tendencia a que un mayor número de transacciones se relacione con mayores pagos. Sin embargo, la relación no es lo suficientemente fuerte como para considerarla significativa de forma general.

**Gráfico de dispersión**

Este gráfico muestra si hay relación directa entre las transacciones y los pagos para cada mes (se puede hacer uno por mes o uno global).

```python
# Crear subgráficas para varios meses (de Enero a Diciembre)
fig, axes = plt.subplots(3, 4, figsize=(18, 12))  # 3 filas, 4 columnas
meses = [f"T{str(i+1).zfill(2)}" for i in range(12)]

# Recorrer cada mes y agregar el gráfico correspondiente
for i, mes in enumerate(meses):
    ax = axes[i//4, i%4]  # Determina la posición de cada subplot
    sns.scatterplot(x=f'Txs_{mes}', y=f'PagoTotal_{mes}', data=data_frame, alpha=0.5, ax=ax)
    ax.set_title(f'Relación entre Transacciones y Pagos Totales - {meses[i]}')
    ax.set_xlabel('Cantidad de Transacciones')
    ax.set_ylabel('Monto de Pagos Totales')
    ax.grid(True)

# Ajuste del layout para que no se sobrepongan los gráficos
plt.tight_layout()
plt.show()
```

---

**Observaciones generales**

- Clientes con pocas transacciones pueden ser **altamente valiosos**, lo que sugiere que deberíamos **profundizar en un análisis de valor por cliente**, más que por volumen de actividad.


**En términos técnicos**

- Una correlación positiva débil implica que, a medida que una variable aumenta, la otra también lo hace, pero de forma poco consistente.

- En la matriz de correlación que calculamos antes, los valores más altos estaban en torno a 0.26 a 0.29, lo cual sí entra en el rango de una correlación positiva débil (generalmente entre 0.1 y 0.3).

- Esto sugiere que la relación no es lineal ni consistente, y que esa leve correlación matemática puede **estar inflada por algunos casos atípicos (outliers)**.

**Conclusión**

Técnicamente sí, hay una correlación positiva débil entre transacciones y pagos, pero:

`Desde una perspectiva de negocio y visual, esa correlación no es útil ni significativa para explicar el comportamiento general de los clientes.`

Sería más relevante agrupar clientes por valor y comportamiento, en lugar de asumir que más transacciones = más ingresos.
