### üîÅ **Metodolog√≠a CRISP-DM**
A continuaci√≥n, se detallan las fases que se utilizar√°n en este proyecto, siguiendo la metodolog√≠a **CRISP-DM (Cross Industry Standard Process for Data Mining)**.

1. Comprensi√≥n del Negocio (Definici√≥n de objetivos)**
2. Comprensi√≥n de los Datos
3. Preparaci√≥n de los Datos
4. Modelado
5. Evaluaci√≥n
6. Despliegue (Posiblemente no considerada)


#### üß≠ **Definici√≥n de fases de la metodolog√≠a Crisp-DM utilizada a lo largo del semestre**

- **Fase 1: Comprensi√≥n del negocio (Business Understanding)**

  En esta etapa, se busca comprender los objetivos y requisitos del negocio desde la perspectiva del problema que se est√° abordando. Se definen los objetivos del proyecto, se establece el marco de trabajo y se desarrolla un plan inicial para alcanzar esos objetivos.

- **Fase 2: Comprensi√≥n de los datos (Data Understanding)**

  Aqu√≠, se recopila y se familiariza con los datos disponibles para el proyecto. Se realiza una exploraci√≥n inicial de los datos para comprender su estructura, calidad, distribuciones y posibles relaciones entre las variables. Esto puede implicar la realizaci√≥n de an√°lisis estad√≠sticos descriptivos, visualizaci√≥n de datos y evaluaci√≥n de la calidad de los datos.

- **Fase 3: Preparaci√≥n de los datos (Data Preparation)**

  Durante esta etapa, se limpian y preparan los datos para el modelado. Esto puede incluir la selecci√≥n de variables relevantes, la transformaci√≥n de datos, la imputaci√≥n de valores faltantes, la codificaci√≥n de variables categ√≥ricas y la creaci√≥n de nuevas caracter√≠sticas derivadas. El objetivo es garantizar que los datos est√©n en un formato adecuado y de alta calidad para el an√°lisis posterior.

- **Fase 4: Modelamiento (Modeling)**

  En esta etapa, se seleccionan y se aplican t√©cnicas de modelado para construir y evaluar modelos predictivos o descriptivos. Se pueden probar varias t√©cnicas y algoritmos de modelado, y se ajustan los par√°metros de los modelos para optimizar su rendimiento. Esta etapa tambi√©n implica la evaluaci√≥n y comparaci√≥n de los modelos para seleccionar el m√°s adecuado seg√∫n los criterios definidos anteriormente.

- **Fase 5: Evaluaci√≥n del modelo (Evaluation)**

  Aqu√≠ se eval√∫an los modelos desarrollados para determinar su idoneidad para abordar los objetivos del negocio. Se utilizan m√©tricas de evaluaci√≥n adecuadas para medir el rendimiento de los modelos, como precisi√≥n, recall, F1-score, entre otras. Adem√°s, se puede realizar una validaci√≥n cruzada o dividir los datos en conjuntos de entrenamiento y prueba para evaluar el rendimiento del modelo en datos no vistos.

- **Fase 6: Despliegue (Deployment)**

  La etapa final implica implementar los resultados del proyecto en el entorno operativo del negocio. Esto puede implicar la integraci√≥n de los modelos en sistemas existentes, la creaci√≥n de informes o tableros interactivos, la automatizaci√≥n de procesos o cualquier otra acci√≥n necesaria para garantizar que los resultados se utilicen de manera efectiva para generar valor comercial.

**1. ¬øQu√© meses presentan mayor o menor facturaci√≥n?**

**Objetivo:** Analizar la facturaci√≥n mensual de los clientes para detectar patrones de estacionalidad o meses con mayores y menores ingresos.

```python
# Convertir las columnas de facturaci√≥n a tipo num√©rico, forzando los errores a NaN
facturacion_columns = ['Fac_T01', 'Fac_T02', 'Fac_T03', 'Fac_T04', 'Fac_T05', 'Fac_T06',
                       'Fac_T07', 'Fac_T08', 'Fac_T09', 'Fac_T10', 'Fac_T11', 'Fac_T12']

# Convertimos las columnas de facturaci√≥n a num√©rico
data_frame[facturacion_columns] = data_frame[facturacion_columns].apply(pd.to_numeric, errors='coerce')

# Calculamos el promedio de facturaci√≥n mensual y redondeamos a enteros
facturacion_mensual = data_frame[facturacion_columns].mean().round(0).astype(int)

# Renombramos los √≠ndices con los nombres de los meses
facturacion_mensual.index = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
                             'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

# Graficamos la facturaci√≥n mensual
plt.figure(figsize=(10, 6))
sns.lineplot(x=facturacion_mensual.index, y=facturacion_mensual.values, marker='o', color='b')
plt.title('Promedio de Facturaci√≥n por Mes')
plt.xlabel('Mes')
plt.ylabel('Facturaci√≥n Promedio')
plt.xticks(rotation=45)
plt.grid(True)

# Agregamos las cifras exactas en los puntos del gr√°fico
for i, valor in enumerate(facturacion_mensual.values):
    plt.text(facturacion_mensual.index[i], valor, str(valor), color='black', ha='center', va='bottom')

# Mostrar el gr√°fico
plt.show()
```

**Objetivo:** Evaluar la relaci√≥n entre las transacciones realizadas por los clientes y los pagos efectivos para ver si estos est√°n directamente relacionados.

**Variables a analizar:**
- **Transacciones por mes:** Txs_T01, Txs_T02, ..., Txs_T12

**Matriz de correlaci√≥n**

Calcularemos la correlaci√≥n entre las transacciones y los pagos para cada mes.

```python
# Definir columnas de transacciones y pagos reales
transacciones = ['Txs_T01', 'Txs_T02', 'Txs_T03', 'Txs_T04', 'Txs_T05', 'Txs_T06',
                 'Txs_T07', 'Txs_T08', 'Txs_T09', 'Txs_T10', 'Txs_T11', 'Txs_T12']

pagos_nacionales = ['PagoNac_T01', 'PagoNac_T02', 'PagoNac_T03', 'PagoNac_T04', 'PagoNac_T05', 'PagoNac_T06',
                    'PagoNac_T07', 'PagoNac_T08', 'PagoNac_T09', 'PagoNac_T10', 'PagoNac_T11', 'PagoNac_T12']

pagos_internacionales = ['PagoInt_T01', 'PagoInt_T02', 'PagoInt_T03', 'PagoInt_T04', 'PagoInt_T05', 'PagoInt_T06',
                         'PagoInt_T07', 'PagoInt_T08', 'PagoInt_T09', 'PagoInt_T10', 'PagoInt_T11', 'PagoInt_T12']

# Crear columnas de pagos totales por mes
for i in range(12):
    mes = f"T{str(i+1).zfill(2)}"
    data_frame[f'PagoTotal_{mes}'] = pd.to_numeric(data_frame[f'PagoNac_{mes}'], errors='coerce') + \
                                     pd.to_numeric(data_frame[f'PagoInt_{mes}'], errors='coerce')

# Convertimos las columnas de transacciones a num√©rico
data_frame[transacciones] = data_frame[transacciones].apply(pd.to_numeric, errors='coerce')

# Seleccionamos las nuevas columnas de pagos totales
pagos_totales = [f'PagoTotal_T{str(i+1).zfill(2)}' for i in range(12)]

# Calculamos la correlaci√≥n entre transacciones y pagos
correlacion = data_frame[transacciones + pagos_totales].corr()

# Extraemos la submatriz de correlaci√≥n entre transacciones y pagos
correlacion_tx_pago = correlacion.loc[transacciones, pagos_totales]

# Visualizamos la matriz de correlaci√≥n
plt.figure(figsize=(10, 8))
sns.heatmap(correlacion_tx_pago, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title('Correlaci√≥n entre Transacciones y Pagos por Mes')
plt.xlabel('Pagos Totales')
plt.ylabel('Transacciones')
plt.show()
```

---

**Observaciones**

---

- La correlaci√≥n m√°s alta es de 0.29, lo que indica una relaci√≥n ligeramente positiva, pero no fuerte.

- Las correlaciones est√°n en torno a 0.26 - 0.29 en el centro del a√±o, lo que sugiere que hay una cierta asociaci√≥n en esos meses.

- No se observan correlaciones negativas fuertes, lo cual tambi√©n es relevante.

**Conclusi√≥n**

Existe una correlaci√≥n positiva d√©bil entre la cantidad de transacciones y los pagos efectivos mensuales. Las correlaciones m√°s altas (m√°ximo de 0.29) se observan en los meses centrales del a√±o, lo que sugiere una ligera tendencia a que un mayor n√∫mero de transacciones se relacione con mayores pagos. Sin embargo, la relaci√≥n no es lo suficientemente fuerte como para considerarla significativa de forma general.
